# Machine Learning Model Implementation
This notebook demonstrates a predictive classification model using scikit-learn. It loads a dataset, preprocesses the data, trains a model, evaluates performance, and makes predictions on new samples.

## 1. Import Required Libraries
We import scikit-learn, pandas, numpy, and plotting utilities for model training and evaluation.

```python
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_openml, load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

## 2. Load and Explore the Dataset
The notebook attempts to load the Spam Base dataset from OpenML. If that is unavailable, it falls back to the breast cancer dataset, which is a reliable classification dataset included with scikit-learn.

```python
def load_dataset():
    try:
        spam = fetch_openml("spambase", version=1, as_frame=True)
        X = spam.data
        y = spam.target.astype(int)
        feature_names = X.columns.tolist()
        description = "Spambase dataset from OpenML"
    except Exception as e:
        print("Could not load the Spambase dataset from OpenML. Falling back to the breast cancer dataset.")
        cancer = load_breast_cancer(as_frame=True)
        X = cancer.data
        y = cancer.target
        feature_names = X.columns.tolist()
        description = "Breast Cancer dataset from scikit-learn"
    return X, y, feature_names, description

X, y, feature_names, dataset_description = load_dataset()
print(dataset_description)
print("Features:", len(feature_names))
print("Target distribution:")
print(y.value_counts())
```

### View a few sample rows and check for missing values

```python
data = X.copy()
data["target"] = y
data.head()
```

```python
print("Missing values in each column:")
print(data.isna().sum().sort_values(ascending=False).head())
print("
Dataset shape:", data.shape)
```

## 3. Preprocess the Data
For numeric features we scale the inputs. If text data were used, this is where vectorization would happen.

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print("Scaled feature sample shape:", X_scaled.shape)
```

## 4. Split Data into Training and Test Sets
We use an 80/20 split for training and evaluation. The random state is fixed for reproducibility.

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)
print("Training set shape:", X_train.shape)
print("Test set shape:", X_test.shape)
```

## 5. Train a Predictive Model with Scikit-learn
We train a logistic regression model, which is a strong baseline for binary classification tasks.

```python
model = LogisticRegression(max_iter=1000, solver="liblinear")
model.fit(X_train, y_train)
print("Model trained successfully.")
```

## 6. Evaluate Model Performance
We evaluate accuracy, precision, recall, F1-score, and display a confusion matrix for the test set.

```python
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, zero_division=0)
recall = recall_score(y_test, y_pred, zero_division=0)
f1 = f1_score(y_test, y_pred, zero_division=0)
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-score: {f1:.4f}")
print("
Classification report:")
print(classification_report(y_test, y_pred, zero_division=0))
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 5))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", cbar=False)
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

## 7. Make Predictions on New Data
We demonstrate predictions on a small set of new samples from the test set.

```python
sample_indices = np.arange(5)
X_new = X_test[sample_indices]
y_new = y_test.iloc[sample_indices] if hasattr(y_test, "iloc") else y_test[sample_indices]
predictions = model.predict(X_new)
print("New sample predictions:")
for i, (true_label, predicted_label) in enumerate(zip(y_new, predictions), start=1):
    print(f"Sample {i}: True={true_label}, Predicted={predicted_label}")
```

## Summary
This notebook implemented a complete scikit-learn classification pipeline: dataset loading, preprocessing, training, evaluation, and prediction. The model performance metrics show how effectively the model can classify new samples.
